```python
class Car:
    def __init__(self):
        self.engine = None
        self.wheels = []
        self.interior = None

class Engine:
    def __init__(self,type,horsepower):
        self.type = type
        self.horsepower = horsepower

class Wheels:
    def __init__(self,size,material):
        self.size = size
        self.material = material

class Interior:
    def __init__(self,seats,color):
        self.seats = seats
        self.color = color
```

  :

```python
class Builder:
    def build_engine(self):
        pass
    
    def build_wheels(self):
        pass
    
    def build_interior(self):
        pass
    
    def get_result(self):
        pass

class CarBuilder(Builder):
    def __init__(self):
        self.car = Car()
    
    def build_engine(self, engine_type, horsepower):
        self.car.engine = Engine(engine_type, horsepower)
    
    def build_wheels(self, size, material):
        self.car.wheels.append(Wheels(size, material))
    
    def build_interior(self, seats, color):
        self.car.interior = Interior(seats, color)
    
    def get_result(self):
        return self.car
```

 :

```python
builder = CarBuilder()
builder.build_engine("V6", 300)
builder.build_wheels(18, "alloy")
builder.build_interior(5, "black")
car = builder.get_result()

print(car.engine.type) # V6
print(car.engine.horsepower) # 300
print(car.wheels[0].size) # 18
print(car.wheels[0].material) # alloy
print(car.interior.seats) # 5
print(car.interior.color) # black
```
